# Load the csv data set from instat 

```python
# Load the Dataset
file_path = 'Games_Henry_McKinney_March_26_2024.xls'  # Update this to your file path
games_data = pd.read_excel(file_path)
```

# Clean and prepare data set 


```python
# Data Cleaning and Preparation
# Correct approach to handle dates in mixed format
games_data['Date'] = pd.to_datetime(games_data['Date'], errors='coerce', dayfirst=True)

# Replace '-' with NaN and convert numeric columns
games_data.replace('-', np.nan, inplace=True)
numeric_columns = ['InStat Index', 'All shifts', 'Goals', 'First assist', 'Second assist', 'Assists',
                   'Faceoffs', 'Faceoffs won', 'Hits', 'Shots', 'Shots on goal', 'Blocked shots', 
                   'Power play shots', 'Short-handed shots', 'Passes to the slot']
for col in numeric_columns:
    games_data[col] = pd.to_numeric(games_data[col], errors='coerce')
```

# Instat Index Montly Average Analysis: 09/23 - 03/24


```python
# Time Series Analysis - Plotting InStat Index Over Time
# Aggregate the InStat Index by month and take the average
games_data['Month'] = games_data['Date'].dt.to_period('M')
monthly_instat_index = games_data.groupby('Month')['InStat Index'].mean()

# Plot the averaged InStat Index over the aggregated months
plt.figure(figsize=(15, 8))
plt.plot(monthly_instat_index.index.astype(str), monthly_instat_index.values, marker='o', linestyle='-', linewidth=2, markersize=5, color='navy')
plt.title('Average Monthly InStat Index Over Time', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Average InStat Index', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()

```

# Instat Index Game by Game Trend: 09/23 - 03/24


```python
# Sort the data by date to ensure the trendline follows the game order
games_data.sort_values('Date', inplace=True)

# Optionally, filter out rows where 'Date' is NaT if dataset might include them
games_data_filtered = games_data.dropna(subset=['Date'])

# Now you can proceed with plotting using games_data_filtered
plt.figure(figsize=(15, 8))
plt.scatter(games_data_filtered['Date'], games_data_filtered['InStat Index'], color='navy', alpha=0.6, label='Game by Game InStat Index')

# Trendline
trendline = games_data_filtered['InStat Index'].rolling(window=5, min_periods=1).mean()
plt.plot(games_data_filtered['Date'], trendline, color='red', linewidth=2, label='Trendline (Rolling Mean)')

# Add opponent tags
for index, row in games_data_filtered.iterrows():
    plt.annotate(row['Opponent'], (row['Date'], row['InStat Index']), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8)

# Formatting the plot
plt.title('Game by Game InStat Index with Trendline and Opponent Tags', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('InStat Index', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.5)
date_form = DateFormatter("%Y-%m-%d")
plt.gca().xaxis.set_major_formatter(date_form)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
print(games_data.columns)  # This will show all the column names in the DataFrame

# Function to convert 'MM:SS' to total minutes
def time_on_ice_to_minutes(time_str):
    if pd.isna(time_str):
        return None  # Return None if the value is NaN
    minutes, seconds = map(int, time_str.split(':'))
    return minutes + seconds / 60

# Apply the conversion to the 'Time on ice' column with the correct case
games_data['Time on Ice (min)'] = games_data['Time on ice'].apply(time_on_ice_to_minutes)
```

# Statistical Analysis and Clutch Factor

```python
# Split the 'Score' column
goals = games_data['Score'].str.split(' : ', expand=True)
games_data['Goals For'] = pd.to_numeric(goals[0])
games_data['Goals Against'] = pd.to_numeric(goals[1])

# Calculate 'Goal Difference'
games_data['Goal Difference'] = games_data['Goals For'] - games_data['Goals Against']

# Calculate the correlation coefficients
for metric in ['Goals', 'Assists', 'InStat Index']:
    # Make sure to only include rows where both Time on Ice and the metric are not NaN
    valid_data = games_data.dropna(subset=['Time on Ice (min)', metric])
    correlation, p_value = stats.pearsonr(valid_data['Time on Ice (min)'], valid_data[metric])
    print(f"\033[1mCorrelation between Time on Ice and {metric}:\033[0m \033[4m{correlation} (p={p_value})\033[0m")

    
print("\n")    





# Define a close game as having a goal difference of 1 or 2
games_data['Is Close Game'] = games_data['Goal Difference'].abs().isin([1, 2])

# Compare average performance metrics in close games to those in other games
close_games_metrics = games_data[games_data['Is Close Game']][['Goals', 'Assists']].mean()
other_games_metrics = games_data[~games_data['Is Close Game']][['Goals', 'Assists']].mean()

# Calculate max lengths for each column
goals_length = max(len("{:.5f}".format(close_games_metrics['Goals'])), len("{:.5f}".format(other_games_metrics['Goals'])))
assists_length = max(len("{:.5f}".format(close_games_metrics['Assists'])), len("{:.5f}".format(other_games_metrics['Assists'])))

# Define title format
title_format = "\033[1m{:>30} {:>45}\033[0m"

# Define row format
row_format = "{:>30} {:>45}"






# Print titles
print(title_format.format("Average metrics in close games:", "Average metrics in other games:"))

# Print rows
print(row_format.format(f"Goals      {close_games_metrics['Goals']:{goals_length}.5f}", f"Goals      {other_games_metrics['Goals']:{goals_length}.5f}"))
print(row_format.format(f"Assists    {close_games_metrics['Assists']:{assists_length}.5f}", f"Assists    {other_games_metrics['Assists']:{assists_length}.5f}"))



# Using prior calculated 'close_games_metrics' and 'other_games_metrics'

# First, create lists for the average counts for close games and other games directly from the metrics
close_games_averages = [close_games_metrics['Goals'], close_games_metrics['Assists']]
other_games_averages = [other_games_metrics['Goals'], other_games_metrics['Assists']]

# Now, create the DataFrame for Seaborn plotting
average_metrics = pd.DataFrame({
    'Metric': ['Goals', 'Assists'] * 2,
    'Average Count': close_games_averages + other_games_averages,
    'Game Type': ['Close Games'] * 2 + ['Other Games'] * 2
})

# Plotting with Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Metric', y='Average Count', hue='Game Type', data=average_metrics)

# Adding labels and title
plt.xlabel('Metrics')
plt.ylabel('Average Count')
plt.title('Average Performance Metrics in Close vs Other Games')
plt.legend(title='Game Type')

# Display the bar chart
plt.show()
```

The output shows us several interesting pieces of information:

1. **Correlation between Time on Ice and Goals**:
   - The correlation coefficient is approximately -0.044, which indicates a very weak inverse relationship between 'Time on Ice' and 'Goals'. However, the correlation is not statistically significant (p=0.892 > 0.05), meaning there's no evidence to conclude that the relationship observed is not due to random chance. Therefore, we can say there's no real correlation between time on ice and the number of goals scored.

2. **Correlation between Time on Ice and Assists**:
   - The correlation coefficient is approximately 0.478, indicating a moderate positive relationship between 'Time on Ice' and 'Assists'. The p-value is 0.071, which is slightly above the common significance level of 0.05. This suggests that there might be a positive correlation between the time spent on ice and the number of assists, but it's not strong enough to be considered statistically significant at the 0.05 level. However, it's close, so this could warrant further investigation or could be significant in a larger sample.

3. **Correlation between Time on Ice and InStat Index**:
   - The correlation coefficient is approximately 0.101, indicating a very weak positive relationship between 'Time on Ice' and 'InStat Index'. The p-value is 0.557, which is not statistically significant. Thus, we cannot confidently state that more time on ice improves the InStat Index based on this data.

4. **Average Metrics in Close Games vs. Other Games**:
   - For close games (where the score difference is 1 or 2), the average goals are 1.0, and assists are 1.4.
   - For other games (not close), the average goals are 1.009, and assists are 1.604.
   - Comparing these metrics, it seems that the number of assists is slightly lower in close games compared to other games, while the number of goals scored doesn't change much. However, the difference here is very small, and without knowing the number of games or performing a statistical test for significance, we cannot say for certain if this is a meaningful difference.

In conclusion, the analysis suggests that time on ice might have a slight relationship with the number of assists but not with goals or overall performance as measured by the InStat Index. Moreover, player performance in terms of goals and assists doesn't appear to significantly improve during close games compared to other games. These interpretations should be taken with caution, as the actual significance of the results would depend on the broader context of the data, the number of observations, the variability within the data, and the statistical power of the tests used.

# Check to see if high faceoff win % associated with high instat index

```python
# It's good practice to explicitly create a copy when working on a subset/slice of a DataFrame to avoid SettingWithCopyWarning warning.
games_data_filtered = games_data_filtered.copy()

# Calculate 'Faceoff Percentage'
games_data_filtered['Faceoff Percentage'] = (games_data_filtered['Faceoffs won'] / games_data_filtered['Faceoffs']) * 100

# Drop rows where any of the two columns are NaN before calculating the correlation
valid_data = games_data_filtered.dropna(subset=['Faceoff Percentage', 'InStat Index'])

# Calculate the correlation coefficient
correlation, p_value = pearsonr(valid_data['Faceoff Percentage'], valid_data['InStat Index'])

# Print out the correlation in bold and the p-value underlined
print(f"\033[1mCorrelation between Faceoff Percentage and InStat Index:\033[0m \033[4m{correlation} (p={p_value})\033[0m")

# Create a heatmap as a visual representation of the relationship
# You can adjust the figure size as needed to ensure it displays well
plt.figure(figsize=(15, 8))
sns.heatmap(valid_data.pivot("Opponent", "InStat Index", "Faceoff Percentage"), annot=True, fmt=".1f", cmap="YlGnBu", linewidths=.5)
plt.title('InStat Index vs. Faceoff Percentage per Game', fontsize=16, fontweight='bold')
plt.xlabel('InStat Index', fontsize=14)
plt.ylabel('Opponent', fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

The analysis found a very weak negative correlation between Faceoff Percentage and InStat Index, and the relationship is not statistically significant (p-value > 0.05). This suggests that there is no meaningful linear association between Henry's performance in the dot and the overall performance as measured by the InStat Index within the dataset examined.

It should be noted that this is a small data set 26 games, there is no limit to faceoffs taken per game (n) and for part of the year Henry played wing. A few tweaks in the setup might actually make for a valuable statistical inferences.

- 1. Faceoff win % and team % time with puck (What % of the player's shift does his team have the puck)

- 2. To build off of #1 Corsi, or xG vs xG against when player on the ice

- 3. A larger data set with a pure center. 100+ games of a someone who played center the entire time.

- 4. As a subsititute to 3, 100+ games with > n faceoffs taken. We could experiment with n, but this gives not only a larger overall data set (100 games minimum) but also a larger data set for individual instances.
